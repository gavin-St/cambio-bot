```python
from card import Card
from game import Game
from game_debug_mode import Game_DebugMode
from player import Player

def main():
    NUMBER_OF_SIMULATIONS = 10000
    game = Game()

    Player1 = Player("Player1", game)
    Player2 = Player("Player2", game)
    Player3 = Player("Player3", game)
    Player4 = Player("Player4", game)
    players = [Player1, Player2, Player3, Player4]

    game.add_players(players)

    # print(game.deck)
    # print(game.players)
    # print(game)
    for i in range(NUMBER_OF_SIMULATIONS):
        print(f"\nGame {i+1}\n\n")
        game.setup_game()
        game.run_game()
        game.reset_game()
    
    print("\n--------------------------------------------------------------")
    print("FINAL standings:")
    for player in players:
        print(f"{player.id}: {player.total_points} points, {player.total_wins} wins.")
    print("--------------------------------------------------------------\n")
    print(f"Total rounds: {game.total_rounds}")

main()
```
